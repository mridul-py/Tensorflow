# Histogram in Matplotlib


```python
from numpy.random import randn #for generating random number datasets (normal distribution)
import matplotlib.pyplot as plt
%matplotlib inline
```

# To take 100 random values


```python
dataset1 = randn(100)
plt.hist(dataset1)
```

# To show the Chart as output

```python
plt.show()
```

# Add Color to histogram


```python
dataset2 = randn(80)
plt.hist(dataset2,color='indianred')
```

```python
plt.show()
```

# Histogram overlapping

```python
plt.hist(dataset1,density=True,alpha=0.5,bins=20)
plt.hist(dataset2,density=True,color='indianred',alpha=0.5,bins=20)
```

```python

plt.show()
```
